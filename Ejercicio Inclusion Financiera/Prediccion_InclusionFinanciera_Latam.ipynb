```python
from sklearn.preprocessing import LabelEncoder
le_bank_account_model  = LabelEncoder()
le_location_type_model  = LabelEncoder()
le_cellphone_access_model  = LabelEncoder()
le_gender_of_respondent_model  = LabelEncoder()
le_relationship_with_head_model  = LabelEncoder()
le_education_level_model  = LabelEncoder()
le_job_type_model  = LabelEncoder()
le_job_type_model  = LabelEncoder()
le_job_type_model  = LabelEncoder()
le_marital_status_model  = LabelEncoder()

def encodeLabels(dataset):   
    
    dataset['bank_account'] = le_bank_account_model.fit_transform(dataset['bank_account'])

    dataset['location_type'] = le_location_type_model.fit_transform(dataset['location_type'])
   
    dataset['cellphone_access'] = le_bank_account_model.fit_transform(dataset['cellphone_access'])
   
    dataset['gender_of_respondent'] = le_gender_of_respondent_model.fit_transform(dataset['gender_of_respondent'])
    
    dataset['relationship_with_head'] = le_relationship_with_head_model.fit_transform(dataset['relationship_with_head'])
    
    dataset['education_level'] = le_education_level_model.fit_transform(dataset['education_level'])
    
    dataset['job_type'] = le_job_type_model.fit_transform(dataset['job_type'])
    
    dataset['marital_status'] = le_marital_status_model.fit_transform(dataset['marital_status'])
    
    return dataset
```

```python
def train_test_for_country(dataset , country):  
    from sklearn.model_selection import train_test_split  
    dataset_country =  dataset[dataset.country == country]
    X_data = dataset.drop(columns=['country' , 'uniqueid' ,'bank_account'])
    y_data = dataset.bank_account
    X_train, X_test, y_train, y_test = train_test_split(X_data , y_data , train_size = 0.8 , stratify= y_data )
    return X_train, X_test, y_train, y_test
    
```

```python
def randonForest(X_train, X_test, y_train, y_test):
    name = 'RandomForest'
    print('-----------------------')
    print(name)
    print('-----------------------')
    from sklearn.model_selection import RandomizedSearchCV
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.metrics import accuracy_score
    
    parametros = {'max_depth' : [5,10,15,20] , 'n_estimators' : [10 ,30 ,50 ,100] }
    rfc = RandomForestClassifier(n_jobs=-1)
    
    random_search = RandomizedSearchCV(rfc , parametros , cv=5 , n_jobs=- 1)
    random_search.fit(X_train, y_train)
    print('Parametros: ',random_search.best_params_)
    print('Accuracy Train: ',random_search.best_score_ )
    rfc = random_search.best_estimator_
    y_pred = rfc.predict(X_test)
    print('Accuracy Test ',accuracy_score(y_pred , y_test) )
    
    return name , random_search.best_estimator_ , accuracy_score(y_pred , y_test) 
    
```

```python
def knn(X_train, X_test, y_train, y_test):
    name = 'KNN'
    print('-----------------------')
    print(name)
    print('-----------------------')
    from sklearn.model_selection import RandomizedSearchCV
    from sklearn.neighbors import KNeighborsClassifier
    from sklearn.metrics import accuracy_score
    
    parametros = {'n_neighbors' : [5,10,15,20,30] , 'weights' : ['uniform' , 'distance'] }
    model = KNeighborsClassifier(n_jobs=-1)
    
    random_search = RandomizedSearchCV(model , parametros , cv=5 , n_jobs=- 1)
    random_search.fit(X_train, y_train)
    print('Parametros: ',random_search.best_params_)
    print('Accuracy Train: ',random_search.best_score_ )
    model = random_search.best_estimator_
    y_pred = model.predict(X_test)
    print('Accuracy Test ',accuracy_score(y_pred , y_test) )
    
    return name , random_search.best_estimator_ , accuracy_score(y_pred , y_test) 
```

```python
def adaBoost(X_train, X_test, y_train, y_test):
    name = 'AdaBoost'
    print('-----------------------')
    print(name)
    print('-----------------------')
    from sklearn.model_selection import RandomizedSearchCV
    from sklearn.ensemble import AdaBoostClassifier
    from sklearn.metrics import accuracy_score
    
    parametros = {'n_estimators' : [20 , 50 , 100 , 200] ,  'learning_rate' : [0.5 , 1  , 1.5]}
    model = AdaBoostClassifier()
    
    random_search = RandomizedSearchCV(model , parametros , cv=5 , n_jobs=- 1)
    random_search.fit(X_train, y_train)
    print('Parametros: ',random_search.best_params_)
    print('Accuracy Train: ',random_search.best_score_ )
    model = random_search.best_estimator_
    y_pred = model.predict(X_test)
    print('Accuracy Test ',accuracy_score(y_pred , y_test) )
    
    return name, random_search.best_estimator_ , accuracy_score(y_pred , y_test) 
```

```python
def neural_NET(X_train, X_test, y_train, y_test):
    name = 'RedNeuronal'
    print('-----------------------')
    print(name)
    print('-----------------------')
    from sklearn.model_selection import RandomizedSearchCV
    from sklearn.neural_network import MLPClassifier
    from sklearn.metrics import accuracy_score
    
    parametros = {'hidden_layer_sizes': [(10,20,10),(20,50,20,10), (2,2,5,10,2)] , 'learning_rate_init' : [0.0001 , 0.001 , 0.005 , 0.01]}
    model = MLPClassifier()
    
    random_search = RandomizedSearchCV(model , parametros , cv=5 , n_jobs=- 1)
    random_search.fit(X_train, y_train)
    print('Parametros: ',random_search.best_params_)
    print('Accuracy Train: ',random_search.best_score_ )
    model = random_search.best_estimator_
    y_pred = model.predict(X_test)
    print('Accuracy Test ',accuracy_score(y_pred , y_test) )
    
    return name , random_search.best_estimator_ , accuracy_score(y_pred , y_test) 
```

```python
def voting(X_train, X_test, y_train, y_test , array_models):
    
    from sklearn.metrics import accuracy_score
    from sklearn.ensemble import VotingClassifier
    
    name = 'Voting'
    print('-----------------------')
    print(name)
    print('-----------------------')
   
    voting_classifier = VotingClassifier(estimators=array_models )
    voting_classifier = voting_classifier.fit(X_train , y_train)
    y_pred = voting_classifier.predict(X_test)
    
    print('Accuracy Test ',accuracy_score(y_pred , y_test) )
    
    return name , voting_classifier , accuracy_score(y_pred , y_test) 
```

```python
def calcular(dataset):
    dict_models = []
    paises = dataset.country.unique()
    print('Paises' , paises )
    for pais in paises :
        print('Datos para pais: ', pais)
        X_train, X_test, y_train, y_test = train_test_for_country(dataset,pais)
        models = []
        for i in range(0 , 5):
            
            if i == 0 :
                name , model ,accuracy = randonForest(X_train, X_test, y_train, y_test )
            elif i == 1 :
                name ,model,accuracy = knn(X_train, X_test, y_train, y_test )
            elif i == 2 :
                name ,model ,accuracy= adaBoost(X_train, X_test, y_train, y_test )
            elif i == 3 :
                name ,model ,accuracy= neural_NET(X_train, X_test, y_train, y_test )
            elif i == 4 :
                name ,model,accuracy = voting(X_train, X_test, y_train, y_test , array_models=models )
            models.append((name , model))
            
            dict_models.append([pais , name , accuracy , model] )
            
    df = pd.DataFrame(dict_models)
    df.columns = ['Pais' , 'nombre' , 'accuracy' , 'modelo']
    return df 
            
            
        
```
